```python
# 1 core data structures
# 1.a
import pandas as pd
s = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
print(s)

# 1.b
data = {
    "Name":["Alice", "Bob", "Charlie"],
    "Age":[22, 23, 24],
}
df = pd.DataFrame(data, index=["a", "b", "c"])
print(df)

# 1.c: Shape, size, dtypes
print(df.shape)
print(df.size)
print(df.dtypes)
```

```python
#2 Data Loading and Saving
csv = pd.read_csv("data.csv")
excel= pd.read_excel("data.xlsx")
json = pd.read_json("data.json")
```

```python
#3 Data Inspection and Understanding (reference by cell 1)
df.head()
df.tail()
df.info()
df.describe()
df.columns
df.index
df.isnull().sum()
```

```python
# 4 Cols and Rows selection
#col selection
print(df["Age"])
df[["Name", "Age"]]
#row selection
#using loc
# print(df.loc[0])
# print(df.loc[0:1]) #range

#using iloc
df.iloc[0]
df.iloc[0:2]

#conditional selection
print(df[df["Age"]>23])
```

```python
# 5 Handling Missing Values
df.isnull()
print(df.isnull().sum())

#remove dups
df.dropna()

#fill missing
df.fillna(0)
df.fillna(method='ffill')
```

```python
# 6 Data Type Conversion

df["Age"] = df["Age"].astype(float)
ddf = df["Age"].astype("category")
ddf
```

```python
# 7 Categorical Data Handling
df["Age"].value_counts()
# df["Age"].cat.categories
# df["Age"].cat.codes
```

```python
# 8 sorting and ranking
df.sort_values("Age")
df.sort_values("Age", ascending=False)
# ranking
df["Rank"] = df["Age"].rank()
print(df["Rank"])
```

```python
# 9 grouping and aggregation
df.groupby("Name")["Age"].mean()
df.groupby("Name")["Age"].max()
df.groupby("Name")["Age"].min()
```

```python
# 10 Merging, Joining, and Concatenation
import pandas as pd
import pandas as pd
s = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
print(s)

# 1.b
data = {
    "Name":["Alice", "Bob", "Charlie"],
    "Age":[22, 23, 24],
}
df = pd.DataFrame(data, index=["a", "b", "c"])
print(df)
pd.merge(df, df, on="Age", how="inner")
```
