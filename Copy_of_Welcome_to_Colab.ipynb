```
# 1 line plots
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [2,4,6,8,10,12]
plt.plot(x, y, marker='o')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Plot")
plt.grid(True)
plt.show()
```

```
# 2 scatter plot

plt.scatter(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")
plt.show()
```

```
# 3 Bar
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 6, 9]
plt.bar(categories, values)
plt.title("Bar Chart")
plt.show()
# horizontal bar
plt.barh(categories, values)
plt.title("Bar-H Chart")
```

```
# 4 histogram
import numpy as np
data = np.random.randn(1000)
plt.hist(data, bins=100)
plt.title("Histogram")
plt.show()
```

```
# 5 Box plot
data = [np.random.randn(100) for _ in range(3)]
plt.boxplot(data)
plt.title("Box plot")
plt.show()
```

```
# 6 subplots
fig, ax = plt.subplots(2, 2)
ax[0, 0].plot([1, 2, 3], [3, 2, 1])
# ax[0, 0].title("Plot")
ax[0, 1].scatter([1, 2, 3], [1, 4, 9])
ax[1, 0].bar(['A', 'B'], [3, 7])
ax[1, 1].hist([1, 2, 2, 3, 3, 3])
plt.tight_layout()
plt.show()
```

```
# 7 figure customization
plt.figure(figsize=(8, 5))
plt.plot([1, 2, 3], [2, 3, 4], linestyle='--', linewidth=2)
plt.title("Customized plots", fontsize=14)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```
# 8 plotting multiple lines
plt.plot([1, 2, 3], [1, 4, 9], label="Square")
plt.plot([1, 2, 3], [1, 8, 27], label="Cube")
plt.legend()
plt.title("Multiple Lines")
plt.show()
```

```
# 9 saving figs

plt.plot([1,2,3], [1, 4, 9])
plt.savefig('output.png', dpi=300)
plt.show()
```
